```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df=pd.read_csv('/content/drive/My Drive/Data/ex1data1.csv')
df.head()
```

```
X=df['X'].values
y=df['y'].values
X=X.reshape((X.shape[0],1))
y=y.reshape((y.shape[0],1))
```

```
plt.scatter(X,y)
plt.title('Linear Regression with one variable')
plt.xlabel('Profit in 10,000s')
plt.ylabel('Population of city in 10,000s')
plt.show()
```

```
def normalise(X):
  return (X-np.mean(X))/np.std(X)
def cost(theta,x,y):
  m=X.shape[0]
  x_one=np.concatenate((np.ones((x.shape[0],1)),x),axis=1)
  cost=1/(2*m)*np.sum((x_one.dot(theta)-y).T.dot(x_one.dot(theta)-y))
  return cost
def train(x,y,alpha,theta):
  x_one=np.concatenate((np.ones((x.shape[0],1)),x),axis=1)
  h=x_one.dot(theta)
  iterations=1500
  m=X.shape[0]
  print(x_one.shape)
  for i in range(0,iterations):
    h=x_one.dot(theta)
    theta=theta-((alpha/m)*(x_one.T.dot((h-y))))
    #print(theta)
  return theta
```

```
theta=np.random.randn(X.shape[1]+1,1)
print(theta)
print(theta.shape)
print(cost(theta,X,y))
X=normalise(X)
y=normalise(y)
theta=train(X,y,0.01,theta)
J=cost(theta,X,y)
print(theta,J)
```

```
x_one=np.concatenate((np.ones((X.shape[0],1)),X),axis=1)
h=x_one.dot(theta)
plt.scatter(X,y)
plt.plot(X,h,color='red')
plt.title('Linear Regression with one variable')
plt.xlabel('Profit in 10,000s')
plt.ylabel('Population of city in 10,000s')
plt.show()
```
